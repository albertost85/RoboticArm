```python
import gym
import rlbench.gym
from stable_baselines3 import A2C
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
env_name = 'reach_target-state-v0'
render_mode = 'rgb_array'
observation_mode = 'state'
policy_mode = 'MlpPolicy'
model_name = 'A2C'
name_model = env_name + '_' + render_mode + '_' + model_name
print(name_model)
```

```python
env = gym.make(env_name, render_mode=render_mode, observation_mode=observation_mode)
model = A2C(policy_mode, env, verbose=1, tensorboard_log="./tb_log_name/")
```

```python
obs = env.reset()
episode_rewards = [0.0]
img = env.render(mode=render_mode)
fig = plt.figure()
im = plt.imshow(img)
plt.show()
```

```python
tip = env.env._robot.arm.get_tip()
```

```python
obs, reward, done, info = env.step(env.action_space.sample())
img = env.render(mode=render_mode)
m = plt.imshow(img)
```

```python

positions = env.env._robot.arm.get_joint_positions()
gripper_pose = np.array(tip.get_pose())
task_low_dim_state = env.task._task.get_low_dim_state()
ee_pose = np.array(obs[22:25])
target_pose = np.array(obs[-3:])

distance = np.sqrt((target_pose[0]-ee_pose[0])**2 +
                   (target_pose[1]-ee_pose[1])**2+(target_pose[2]-ee_pose[2])**2)

# reward = np.tanh(1/(distance+1e-5))
reward = np.exp(-1*distance)
print(distance)
print(reward)
```
